# Modelo de Lotka Volterra para poblaciones en competencia por recursos

Las ecuaciones diferenciales de este modelo son:
$$
\frac{d x}{d t} = x \left( 1 - x - \alpha y \right) \\
\frac{d y}{d t} = \rho y  \left( 1 - y - \beta x\right)
$$

En este modelo siempre existen 3 puntos fijos $(0, 0)$, $(0, 1)$ y $(1, 0)$. Puede existir un cuarto punto fijo $\left( \frac{1 - \alpha}{1 - \alpha \beta}, \frac{1 - \beta}{1 - \alpha \beta} \right)$, pero s√≥lo si $\alpha, \beta < 1$ o si $\alpha, \beta > 1$.

En cuanto a la estabilidad de los puntos fijos:

* El punto fijo $(0, 0)$ es siempre inestable.

* El punto fijo $(0, 1)$ Es estable si $\alpha > 1$

* El punto fijo $(1, 0)$ Es estable si $\beta > 1$

* El punto fijo $\left( \frac{1 - \alpha}{1 - \alpha \beta}, \frac{1 - \beta}{1 - \alpha \beta} \right)$ Es estable si $\alpha, \beta < 1$



```python
def modelo(p, t, alpha, beta, rho):
    x, y = p
    dx = x * (1 - x - alpha * y)
    dy = rho * y * (1 - y - beta * x)
    return dx, dy
```
