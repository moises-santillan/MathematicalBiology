```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def modelODE_Alle(N, t, r, K, A):
    return r*N*(1-N/K)*(N/A-1)
```

```python
r = 1
K = 1
A=0.3
t = 0

N = np.linspace(0, 1.1, 200)

fig, ax =  plt.subplots()

f = modelODE_Alle(N, t, r, K, A)
ax.plot(N, f)

ax.axhline(y = 0, color = 'k', linestyle = '--') 

ax.set_xlabel(r"Population Size ($N$)", fontsize=16)
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
```

```python
t = np.linspace(0, 20, 500)
r = 1
K = 1
A = 0.3

fig, ax =  plt.subplots()

N0 = .1
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .15
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .25
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .29
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .299
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .3
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .301
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .31
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .5
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .8
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 1.1
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 1.5
N = odeint(modelODE_Alle, N0, t, args=(r, K, A))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

ax.set_xlabel("Time (t)", fontsize=16)
ax.set_ylabel("Population Size (N)", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
ax.legend(fontsize=14)
```

```python
def modelODE(N, t, r, K):
    return r*N*(1-N/K)

def modelODE_W(N, t, r, K):
    return r * N**2 * (1-N/K)
```

```python
r = 1
K = 1
t = 0

N = np.linspace(0, 1.1, 200)

fig, ax =  plt.subplots()

f = modelODE(N, t, r, K)
ax.plot(N, f, label=r"Logistic")

f = modelODE_W(N, t, r, K)
ax.plot(N, f, label=r"Weak All")


ax.axhline(y = 0, color = 'k', linestyle = '--') 

ax.set_xlabel(r"Population Size ($N$)", fontsize=16)
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
ax.legend(fontsize=14)
```
