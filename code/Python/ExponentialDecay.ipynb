```python
Nsteps = 1000
Nexps = 100
Pinit = 100000
dp = 0.01
Pop = np.zeros((Nsteps, Nexps))
Pop[0, :] = np.ones(Nexps) * Pinit

for i in range(Nexps):
    x = np.ones(Pinit)
    for j in range(1, Nsteps):
        y = np.random.rand(Pinit)
        idx = np.where(y < dp)[0]
        x[idx] = 0
        Pop[j, i] = np.sum(x)


for i in range(Nexps):
    plt.plot(Pop[:, i])
plt.yscale('log')
```
