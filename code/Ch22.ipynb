```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Define the differential equation
def dx_dt(x, t, gamma, K):
    return gamma*((K + 1) / (K + x) - x)
```

```python
# Time points where solution is computed
t = np.linspace(0, 5, 100)

# Initial condition
x0 = 2
# Parameter gamma
gamma = 1

# Parameter K
K = 2
# Solve the differential equation
x = odeint(dx_dt, x0, t, args=(gamma, K))
# Plot the solution
plt.plot(t, x, label=fr'$K = {K}$')

# Parameter K
K = 1
# Solve the differential equation
x = odeint(dx_dt, x0, t, args=(gamma, K))
# Plot the solution
plt.plot(t, x, label=fr'$K = {K}$')

# Parameter K
K = .5
# Solve the differential equation
x = odeint(dx_dt, x0, t, args=(gamma, K))
# Plot the solution
plt.plot(t, x, label=fr'$K = {K}$')

# Parameter K
K = .2
# Solve the differential equation
x = odeint(dx_dt, x0, t, args=(gamma, K))
# Plot the solution
plt.plot(t, x, label=fr'$K = {K}$')

# Parameter K
K = .1
# Solve the differential equation
x = odeint(dx_dt, x0, t, args=(gamma, K))
# Plot the solution
plt.plot(t, x, label=fr'$K = {K}$')


plt.xlabel('Time')
plt.ylabel('x(t)')
plt.title(r'Solution of $dx/dt = \gamma((K+1)/(K+x) - x$)')
plt.legend()
```
