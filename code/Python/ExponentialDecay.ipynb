```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
Pinit = 100000
dp = 0.01

x = np.zeros((N, Pinit))
x[0, :] = 1
for i in range(1, N):
    y = np.random.rand(Pinit)
    idx = np.where(y < dp)[0]
    x[i, :] = x[i-1, :]
    x[i, idx] = 0
Pop = np.sum(x, axis=1)

plt.plot(Pop)


x = np.zeros((N, Pinit))
x[0, :] = 1
for i in range(1, N):
    y = np.random.rand(Pinit)
    idx = np.where(y < dp)[0]
    x[i, :] = x[i-1, :]
    x[i, idx] = 0
Pop = np.sum(x, axis=1)

plt.plot(Pop)


x = np.zeros((N, Pinit))
x[0, :] = 1
for i in range(1, N):
    y = np.random.rand(Pinit)
    idx = np.where(y < dp)[0]
    x[i, :] = x[i-1, :]
    x[i, idx] = 0
Pop = np.sum(x, axis=1)

plt.plot(Pop)


plt.yscale('log')
```
