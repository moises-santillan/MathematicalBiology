```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def Jin(t, ti, tf, h):
    if (t >= ti) and (t <= tf):
        aux = h
    else:
        aux = 0
    return aux

def model(u, t, ti, tf, h, mu):
    return Jin(t, ti, tf, h) - mu * u
```

```python
mu = np.log(2)/20.
t = np.linspace(0, 120, 2000)

ti = 10
tf = 20
h = 10
u = odeint(model, 0, t, args=(ti, tf, h, mu))
plt.plot(t, u)

ti = 10
tf = 30
h = 5
u = odeint(model, 0, t, args=(ti, tf, h, mu))
plt.plot(t, u)

ti = 10
tf = 50
h = 2.5
u = odeint(model, 0, t, args=(ti, tf, h, mu))
plt.plot(t, u)

```

```python
def Jin_per(t, ti, T, dt, h):
    n = (t-ti)//T
    if (t - ti - n*T) <= dt:
        aux = h
    else:
        aux = 0
    return aux

def model_per(u, t, ti, T, dt, h, mu):
    return Jin_per(t, ti, T, dt, h) - mu * u
```

```python
ti = 10
T = 120
dt = 40
h = 2.5
mu = np.log(2)/60

t = np.linspace(0, 600, 6000)

u = odeint(model_per, 0, t, args=(ti, T, dt, h, mu))

plt.plot(t,u)
```

```python
def Jin(t, ti, tf, h):
    if (t >= ti) and (t <= tf):
        aux = h
    else:
        aux = 0
    return aux

def two_comp_model(u, t, ti, tf, h, mu, gamma):
    du = np.zeros(2)
    du[0] = Jin(t, ti, tf, h) - mu * u[0] - gamma * (u[0] - u[1])
    du[1] = gamma * (u[0] - u[1])
    return du
```

```python
mu = np.log(2)/20.
t = np.linspace(0, 120, 2000)

gamma = 0
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])


gamma = .1*mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])

gamma = mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])

gamma = 1000*mu
ti = 10
tf = 30
h = 5
u = odeint(two_comp_model, [0, 0], t, args=(ti, tf, h, mu, gamma))
plt.plot(t, u[:,0])

```
