```python
Nsteps = 1000
Nexps = 100
Pinit = 100000
dp = 0.01
Pop = np.zeros((Nsteps, Nexps))
Pop[0, :] = np.ones(Nexps) * Pinit

for i in range(Nexps):
    x = np.ones(Pinit)
    for j in range(1, Nsteps):
        y = np.random.rand(Pinit)
        idx = np.where(y < dp)[0]
        x[idx] = 0
        Pop[j, i] = np.sum(x)


for i in range(Nexps):
    plt.plot(Pop[:, i])
plt.yscale('log')
```

```python
endTime = 1500
steps = 201
Nexps = 1000
Pinit = 100000
timeLife = 100

times = np.linspace(0, endTime, steps)
Pop = np.zeros((steps, Nexps))

for i in range(Nexps):
    ages = expon.rvs(size=Pinit, scale=timeLife)
    for j in range(steps):
        idx = np.where(times[j] < ages)[0]
        Pop[j, i] = len(idx)

Texp = np.zeros((steps, Nexps))
for i in range(Nexps):
    Texp[:, i] = timeLife * np.log(Pinit/Pop[:, i])
```

```python
for i in range(Nexps):
    plt.plot(times, Pop[:, i])
plt.yscale('log')
```

```python
for i in range(Nexps):
    plt.plot(times, Texp[:, i])
#plt.yscale('log')
```

```python
Pavg=np.mean(Pop, axis=1)
Pstd=np.std(Pop, axis=1)
Pcv=Pstd/Pavg

Tavg=np.mean(Texp, axis=1)
Tstd=np.std(Texp, axis=1)
Tcv=Tstd/Tavg
```
