```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.signal import square, sawtooth
```

```python
# Define models for different input functions f(t)
def model(x, t, tau):
    # f(t) = 1 (constant input)
    return  (1 - x)/tau

def modelSin(x, t, tau):
    # f(t) = sin(2πt) (sinusoidal input)
    return  (np.sin(2*np.pi*t) - x)/tau

def modelSquare(x, t, tau):
    # f(t) = square(2πt) (square wave input)
    return  (square(2*np.pi*t) - x)/tau

def modelSawtooth(x, t, tau):
    # f(t) = sawtooth(2πt) (sawtooth wave input)
    return  (sawtooth(2*np.pi*t) - x)/tau
```

```python
# Solve and plot for constant input
x0 = 0.0     # Initial condition
t = np.linspace(0, 15, 500)  # Time points

# Try different values of tau (rate constant)
for tau in [0.1, 0.5, 1, 2.0, 5]:
    x = odeint(model, x0, t, args=(tau,))
    plt.plot(t, x, label=rf'$\tau$ = {tau}')

plt.title(r'Solution of $dx/dt = (1 - x)/\tau$')
plt.xlabel(r'Time ($t$)')
plt.ylabel(r'$x(t)$')
plt.legend()
```

```python
# Solve and plot for sinusoidal input
x0 = 0.0     # Initial condition
t = np.linspace(0, 5, 500)  # Time points

for tau in [0.02, 0.1, 0.2, 0.5, 1]:
    x = odeint(modelSin, x0, t, args=(tau,))
    plt.plot(t, x, label=rf'$\tau$ = {tau}')

plt.title(r'Solution of $dx/dt = (\sin(2\pi t) - x)/\tau$')
plt.xlabel(r'Time ($t$)')
plt.ylabel(r'$x(t)$')
plt.legend()
```

```python
# Solve and plot for square wave input
x0 = 0.0     # Initial condition
t = np.linspace(0, 5, 5000)  # Time points

for tau in [0.02, 0.1, 0.2, 0.5, 1]:
    x = odeint(modelSquare, x0, t, args=(tau,))
    plt.plot(t, x, label=rf'$\tau$ = {tau}')

plt.title(r'Solution of $dx/dt = (\text{square}(2\pi t) - x)/\tau$')
plt.xlabel(r'Time ($t$)')
plt.ylabel(r'$x(t)$')
plt.legend()
```

```python
# Solve and plot for sawtooth wave input
x0 = 0.0     # Initial condition
t = np.linspace(0, 5, 5000)  # Time points

for tau in [0.02, 0.1, 0.2, 0.5, 1]:
    x = odeint(modelSawtooth, x0, t, args=(tau,))
    plt.plot(t, x, label=rf'$\tau$ = {tau}')

plt.title(r'Solution of $dx/dt = (\text{sawtooth}(2\pi t) - x)/\tau$')
plt.xlabel(r'Time ($t$)')
plt.ylabel(r'$x(t)$')
plt.legend()
```
