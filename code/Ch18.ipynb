```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model(u, t, mu):
    x, y = u
    dxdt = y
    dydt = -x - mu*y
    return [dxdt, dydt]
```

```python
t = np.linspace(0, 10*np.pi, 1000)
u0 = [1, 0]

mu = 0
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

mu = 0.1
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

mu = 0.5
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

mu = 1
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

mu = 2
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

mu = 4
u = odeint(model, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')


plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
```

```python
def model2(u, t, mu):
    x, y = u
    dxdt = y
    dydt = -x - mu*(x**2-1)*y
    return [dxdt, dydt]
```

```python
t = np.linspace(0, 500*np.pi, 1000)
u0 = [1, 0]

mu = 100
u = odeint(model2, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
```
