```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def SIR_Full(u, t, alpha, beta, gamma, delta, mu):

    du = np.zeros(3)
    du[0] = alpha - beta*u[0]*u[1] - mu*u[0]
    du[1] = beta * u[0]*u[1] - (mu + gamma + delta) * u[1]
    du[2] = gamma * u[1] - mu * u[2]
    return du
```

```python
R0 = 2
mu = 1. / (75*52)
gamma = 1000*mu 
delta = 0.0 
alpha = mu
beta = R0 * mu * (mu + gamma + delta) / alpha

t = np.linspace(0, 1040, 40001)

u = odeint(SIR_Full, [.89, .01, 0], t, args=(alpha, beta, gamma, delta, mu))

plt.plot(t/52., u[:,0], color='C0', label="S")
plt.plot(t/52., u[:,1], color='C1', label="I")
plt.plot(t/52., u[:,2], color='C2', label="R")
plt.plot(t/52., np.sum(u, axis=1), color='C3', label="M")
plt.legend()
plt.xlabel("t (years)")
plt.ylabel("Population size")
```

```python
plt.plot(t, alpha*np.ones(len(t)), label="births")
plt.plot(t, beta*u[:,0]*u[:,1], label="infections")
plt.plot(t, gamma*u[:,1], label="recoveries")
plt.plot(t, mu*np.sum(u, axis=1), label="deaths")
plt.legend()
```

```python
def SIR(u, t, R0, rho):
    du = np.zeros(2)
    du[0] = 1 - R0*u[0]*u[1] - u[0]
    du[1] = rho * (R0*u[0]*u[1] - u[1])
    return du

def SIR_CP(u, t, R0):
    du = np.zeros(2)
    du[0] = -R0*u[0]*u[1]
    du[1] = R0*u[0]*u[1] - u[1]
    return du
```

# Modelo SIR normal

$$
    R_0 = \frac{\alpha \beta}{\mu(\mu + \gamma + \delta)}, \quad \rho = \frac{\mu + \gamma + \delta}{\mu}
$$
