```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def modelSolution(t, N0, r, K):
    return 1/(1/K + (1/N0 - 1/K)*np.exp(-r*t))
```

```python
t = np.linspace(0, 10, 500)
r = 2
K = 1

fig, ax =  plt.subplots()

N0 = .01
N = modelSolution(t, N0, r, K)
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .1
N = modelSolution(t, N0, r, K)
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .5
N = modelSolution(t, N0, r, K)
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 1.5
N = modelSolution(t, N0, r, K)
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 2
N = modelSolution(t, N0, r, K)
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

ax.set_xlabel("Time (t)", fontsize=16)
ax.set_ylabel("Population Size (N)", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
ax.legend(fontsize=14)
```

```python
def modelODE(N, t, r, K):
    return r*N*(1-N/K)
```

```python
t = np.linspace(0, 10, 200)
r = 2
K = 1

fig, ax =  plt.subplots()

N0 = .0
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .01
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .1
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = .5
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 1.5
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")

N0 = 2
N = odeint(modelODE, N0, t, args=(r, K))
ax.plot(t, N, label=rf"$N_0 =$ {N0}")



ax.set_xlabel("Time (t)", fontsize=16)
ax.set_ylabel("Population Size (N)", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
ax.legend(fontsize=14)
```

```python
def modelODE_bis(N, t, r, K, gamma):
    return r*N*(1-N/K) - gamma*N
```

```python
r = 1
K = 1
t = 0

N = np.linspace(0, 1.1, 200)

fig, ax =  plt.subplots()

gamma=0
f = modelODE_bis(N, t, r, K, gamma)
ax.plot(N, f, label=rf"$\gamma =$ {gamma}")

gamma=0.2
f = modelODE_bis(N, t, r, K, gamma)
ax.plot(N, f, label=rf"$\gamma =$ {gamma}")

gamma=0.5
f = modelODE_bis(N, t, r, K, gamma)
ax.plot(N, f, label=rf"$\gamma =$ {gamma}")

gamma=1
f = modelODE_bis(N, t, r, K, gamma)
ax.plot(N, f, label=rf"$\gamma =$ {gamma}")

gamma=1.5
f = modelODE_bis(N, t, r, K, gamma)
ax.plot(N, f, label=rf"$\gamma =$ {gamma}")

ax.axhline(y = 0, color = 'k', linestyle = '--') 

ax.set_xlabel(r"Population Size ($N$)", fontsize=16)
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.tick_params(axis='both', labelsize=14)
ax.legend(fontsize=14)
```
