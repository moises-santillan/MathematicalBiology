```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def oscillator(u, t, omega, eta):
    x, v = u
    dx = v
    dv = -omega**2 * x - eta * v
    return [dx, dv]
```

```python
t = np.linspace(0, 30, 3001)
omega = 1
eta = 1.5

u0 = [1, 0]

u = odeint(oscillator, u0, t, args=(omega, eta))
```

```python
plt.plot(t, u[:, 0])
plt.plot(t, u[:, 1])
```

```python
t = np.linspace(0, 30, 3001)
omega = 1
eta = 1.

u0 = [1, 0]
u = odeint(oscillator, u0, t, args=(omega, eta))
plt.plot(u[:, 0], u[:, 1], color="C0")

u0 = [-1, 0]
u = odeint(oscillator, u0, t, args=(omega, eta))
plt.plot(u[:, 0], u[:, 1], color="C0")

u0 = [0, 1]
u = odeint(oscillator, u0, t, args=(omega, eta))
plt.plot(u[:, 0], u[:, 1], color="C0")

u0 = [0, -1]
u = odeint(oscillator, u0, t, args=(omega, eta))
plt.plot(u[:, 0], u[:, 1], color="C0")

vx = np.linspace(-1, 1, 100)
vy = (-eta + np.sqrt(eta**2 - 4))/2 * vx
plt.plot(vx, vy, color="C1")

vy = (-eta - np.sqrt(eta**2 - 4))/2 * vx
plt.plot(vx, vy, color="C1")
```
