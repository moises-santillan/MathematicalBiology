```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model1(u, t, mu):
    x, y = u
    dxdt = y
    dydt = -x - mu*(x**2-1)*y
    return [dxdt, dydt]
```

```python
t = np.linspace(0, 500*np.pi, 1000)
u0 = [1, 0]

mu = 100
u = odeint(model1, u0, t, args=(mu,))
plt.plot(t, u[:,0], label=rf'$\mu = {mu}$')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
```

```python
def model2(u, t, mu):
    x, y = u
    dxdt = mu * (x - x**3/3 - y)
    dydt = x / mu
    return [dxdt, dydt]
```

```python
t = np.linspace(0, 20*np.pi, 1000)
u0 = [1, 0]

mu = 10
u = odeint(model2, u0, t, args=(mu,))
plt.title(rf'$\mu = {mu}$')
plt.plot(t, u[:,0], label='x(t)')
plt.plot(t, u[:,1], label='y(t)')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
```

```python
#plot of the phase space trajectory
plt.plot(u[:,0], u[:,1])
plt.xlabel('x')
plt.ylabel('y')
plt.title(rf'$\mu = {mu}$')
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(6, 6))

# Set the labels for the axes
ax.set_xlabel(r'$x$', fontsize=16)
ax.set_ylabel(r'$y$', fontsize=16)

# Create the x-nullcline
x = np.linspace(-2.2, 2.2, 400)
y_nullcline = x - (x**3) / 3
ax.plot(x, y_nullcline, color='C0', linewidth=2, label=r'$x$-Nullcline')

# Create the y-nullcline
ax.axvline(0, color='C1', linewidth=2, label=r'$y$-Nullcline')

# Mark the fixed point
ax.scatter([0], [0], color='C2', s=50, label='Fixed Point')  # Fixed point

ax.legend(fontsize=14, loc='lower left')
ax.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Set up the figure
fig, ax = plt.subplots(figsize=(6, 6))

# Define the domain for x and y
x = np.linspace(-15, 15, 30)
y = np.linspace(-100, 100, 30)

# Create a meshgrid for the vector field
X, Y = np.meshgrid(x, y)

# Define the vector field based on the given dynamics
mu = 10
U = mu * (X - (X**3 / 3) - Y)
V = X / mu

# Create the quiver plot
ax.quiver(X, Y, U, V, color='olive', scale=7e4)

# Draw the axes
ax.axvline(0, color='C1', lw=2, label='$y-Nullcline$')

# Plot the nullcline
nullcline_x = np.linspace(-6.7, 6.7, 400)
nullcline_y = nullcline_x - (nullcline_x**3 / 3)
ax.plot(nullcline_x, nullcline_y, color='C0', lw=2, label='x-Nullcline')

# Set the limits and labels
ax.set_title(rf'$\mu={mu}$', fontsize=16)
ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$y$', fontsize=16)
ax.grid(False)
ax.tick_params(axis='both', which='major', labelsize=14)

# Show the plot
plt.legend(fontsize=14)
```
